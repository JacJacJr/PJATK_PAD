```python
orders = pd.read_csv('orders.csv')

#orders.info()
#orders.head()
#1
orders['order_date'] = pd.to_datetime(orders['order_date'])
#w momencie w którym dodaję format='%Y%m%d' wyskakuje ValueError: time data '2016/09/04' does not match format '%Y%m%d' (match)
orders.info()
orders.head()

#2
orders['tshirt_category'].unique()
#array(['White T-Shirt M', 'White T-Shirt F', 'Wh Tshirt F',
#       'Black T-Shirt F', 'Hoodie', 'Black T-Shirt M', 'Tennis Shirt',
#       'Wh Tshirt M', 'Bl Tshirt F', 'Bl Tshirt M'], dtype=object)

#3
orders['tshirt_category'].str.strip()
orders['tshirt_category'].replace({'Wh Tshirt M':'White T-Shirt M', 'Wh Tshirt F':'White T-Shirt F', 'Bl Tshirt F':'Black T-Shirt M', 'Bl Tshirt M':'Black T-Shirt M'}, inplace=True)
orders['tshirt_category'].unique()

#4
orders[["tshirt_color", "tshirt_type", "size"]] = orders['tshirt_category'].str.split(' ', -1, expand=True)
#Jak dodac warunek do petli
#orders[["tshirt_color", "tshirt_type", "size"]] = orders['tshirt_category'].str.apply(lambda x: if orders['tshirt_category'] in 'Hoodie': orders['tshirt_category'].split(' ', -1, expand=True))
orders.sample(20)

#5
from datetime import datetime
d_from = '2014.12.31'
d_to = '2016.08.02'
d_from = datetime.strptime(d_from, '%Y.%m.%d')
d_to = datetime.strptime(d_to, '%Y.%m.%d')
print(f'{type(d_from)} !!!!!!!!!!!!!!!!!!!!!!')

print(f'\nOrders between {d_from} and {d_to}\n')
#nie rozumiem dlaczego nie działa warunek - osobno działają
orders_between = orders[(orders['order_date'] > d_from) & ([orders['order_date']] < d_to)]
#orders_between = orders[(orders['order_date'] > d_to)]
orders_between.sample(20)
```
